```python
import requests

id = '21173'

s = requests.Session()
cookies = dict(I18N_LANGUAGE='zh')


location_url = 'https://ebird.org/tripreport-internal/v1/locations/{}'
taxon_url = 'https://ebird.org/tripreport-internal/v1/taxon-list/{}'
detail_url = 'https://ebird.org/tripreport-internal/v1/taxon-detail/{}/{}'

ret = s.get(location_url.format(id), cookies=cookies)
location=ret.json()

print(location)
```

```python

ret = s.get(taxon_url.format(id), cookies=cookies)
taxon = ret.json()
print(taxon)
```

```python
for i in taxon:
    r=s.get(detail_url.format(id, i['speciesCode']))
    detail=r.json()
    print(detail)
```
