```python
data=[]
fieldname =['date','type','name','starttime','endtime','location','distance','birds','p1','p2','p3','p4', 'ebird','cancel','cancel_help']
for i in range (3, df.shape[0]):
    # 例行
    if type(df.iat[i, 0]) != str and not pd.isnull(df.iat[i, 2]) and not pd.isnull(df.iat[i, 9]) and  df.iat[i,9].strip() !='會員大會(不排)':
        print(df.iat[i, 0],  df.iat[i, 2],  df.iat[i, 3],  df.iat[i, 6],df.iat[i, 7], df.iat[i, 9], df.iat[i, 10], df.iat[i, 11], df.iat[i, 12])

        data.append({
            'date': (df.iat[i, 0]+pd.DateOffset(years=1)).strftime('%Y/%m/%d'),
            'type': '例行',
            'name': df.iat[i, 2].strip(),
            'starttime': df.iat[i, 3].strftime('%H:%M'),
            'endtime': '12:00',
            'location': df.iat[i, 6],
            'distance': df.iat[i, 7],
            'birds': df.iat[i, 8],
            'p1': df.iat[i, 9].strip(),
            'p2': df.iat[i, 10].strip(),
            'p3': df.iat[i, 11].strip(),
            'p4': df.iat[i, 12].strip(),

        })

        
print(data)
```
